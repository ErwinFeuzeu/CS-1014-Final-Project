# Question 1

**How do the level of cannabis use for 12-17 and 17-25 year olds compare for California and New York?**

Answering this question involves comparing the scatter plots of cannabis use in the past month, for both 12-17 year olds and 17-25 year olds.

## Code for question

In the following code iteration is used to examine the drug reports. The name of the state where the drug report is recorded is obtained using the key "State". If the reported drug report is in the state of interest, California or New York, it's age group, 18-25 or 12-17, will be appended into the corresponding list initialized earlier. The age group is obtained using the keys, "Rates", "Marijuana", and "Used Past Month". After the iteration, a scatter plot is created for each list. Titles and labels are added to explain the content of the visualization


```python

CA_Q1_P1_list = []
CA_Q1_P2_list = []
NY_Q1_P1_list = []
NY_Q1_P2_list = []

for drug in drug_list:
    if drug["State"] == "California":
        CA_Q1_P1_list.append(drug["Rates"]["Marijuana"]["Used Past Month"]["18-25"])
    if drug["State"] == "California":
        CA_Q1_P2_list.append(drug["Rates"]["Marijuana"]["Used Past Month"]["12-17"])
    if drug["State"] == "New York":
        NY_Q1_P1_list.append(drug["Rates"]["Marijuana"]["Used Past Month"]["18-25"])
    if drug["State"] == "New York":
        NY_Q1_P2_list.append(drug["Rates"]["Marijuana"]["Used Past Month"]["12-17"])


plt.scatter(CA_Q1_P2_list, CA_Q1_P1_list)
plt.xlabel("% of people 12-17 that has used Marijuana in the past month")          
plt.ylabel("% of people 18-25 that has used Marijuana in the past month")          
plt.title("Age comparision in California") 
plt.show()
plt.scatter(NY_Q1_P2_list, NY_Q1_P1_list)
plt.xlabel("% of people 12-17 that has used Marijuana in the past month")          
plt.ylabel("% of people 18-25 that has used Marijuana in the past month")          
plt.title("Age comparision in New York") 
plt.show()

        

        
```

## Answer to Question 1

The scatter plots above show that while there does appear to be a  positive correlation between the age groups in California, New York does appear to have to same case. You can see how that while accounting for a few outliers the scatter plot for California is a positively slope line but for New York no line can by mapped out as the points are scattered except for the cluster in the top right. Also it is apparent that in both states the percentage is greater for the 18-25 age group.

# Question 2

Another comparision of California and New York is whether there is a difference in how marijuana has been trending. This is captured by the following question:

**What is the trend of cannabis use for 17-25 year olds in California and New York?**

Answering this question requires comparing the line graphs of both states for marijuna use in the past month over time. 

## Code for question 2

The code in the following cell produces a line graph for each state of 17-25 marijuna use in the past month over time. Because we already created a list of this age group's marijuna use in the past month for bothe list, we just need to create a list for the years of both states. The code used iteration to create these two list, as it checked if the report was present in the two states of interest and if so would append the year to the corresponding state list. The year of the report was accessed with the "Year" key.



```python
CA_year = []
NY_year = []
for drug in drug_list:
    if drug["State"] == "California":
        CA_year.append(drug["Year"])
    if drug["State"] == "New York":
        NY_year.append(drug["Year"])
plt.plot(CA_year, CA_Q1_P1_list)
plt.xlabel("Year")          
plt.ylabel("% of people 18-25 that has used Marijuana in the past month")          
plt.title("Cannabis trends in California") 
plt.show()

plt.plot(NY_year, NY_Q1_P1_list)
plt.xlabel("Year")          
plt.ylabel("% of people 18-25 that has used Marijuana in the past month")          
plt.title("Cannabis trends in New York") 
plt.show()
```

## Answer to question 2

The line plots above show that California and New York have different trends in cannabis use over the indicated time span. Cannabis use in California has been postitively trending over time for this age group as indicated by the positive slope of its line plot. New York on the other hand has been trending downwards in terms of cannabis use over time as shown by its line plot's steep decline around the year 2009. The difference in trend between the two states could be caused by the fact the cannabis is only decriminilized in New York[1].

# Question 3

Another way of comparing cannabis activity in California and New York is by asking:

**What is the relationship between perceptions of risk and cannabis use in the past month for 17-25 year olds in California and New York?**

Answering this question requires comparing the scatter plots both age groups in both states. 

## Code for question 3

The code in the following cell produces a scatter plot for each state of *17-25 Marijuna use in the past month* vs *17-25 Perceptions of Risk*. Because we already established a list of this age group's marijuna use in the past month for both states, we just need to create a list for the Perceptions of Risk of this age group for California and New York. The code used iteration to create these two list by filtering the information by state before appending to the two seperate list for  risk. This age group's risk values are obtained using the keys *Rates*, *Marijuana*, *Perceptions of Risk*, and *17-25*.



```python
CA_risk = []
NY_risk = []

for drug in drug_list:
    if drug["State"] == "California":
        CA_risk.append(drug["Rates"]["Marijuana"]["Perceptions of Risk"]["18-25"])
    if drug["State"] == "New York":
        NY_risk.append(drug["Rates"]["Marijuana"]["Perceptions of Risk"]["18-25"])
        
plt.scatter(CA_risk, CA_Q1_P1_list)
plt.xlabel("% of the people that sees risk in Smoking Marijuana, among this age group.")          
plt.ylabel("% of people 18-25 that has used Marijuana in the past month")          
plt.title("Risk vs Use comparision in California") 
plt.show()
plt.scatter(NY_risk, NY_Q1_P1_list)
plt.xlabel("% of the people that sees risk in Smoking Marijuana, among this age group.")          
plt.ylabel("% of people 18-25 that has used Marijuana in the past month")          
plt.title("Risk vs Use comparision in New York") 
plt.show()
    
```

## Answer to question 3

The scatter plots below show that once again California and New York appear to have differing trends. For California there appears to be a negative correlation meaning that the greater the risk found in smoking cannabis the less likely this age group is to participate. This thought process would appear to be the "common sense" one as most people would prefer to engage in activity with lower risk to their health so it's fair to say a trend like this was expected. However, in New York, the trend is opposite as the correlation is postitive. That means that the greater the risk is the more likely a person of this age group in New York will be to engage in smoking cannabis.
